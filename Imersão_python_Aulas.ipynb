Aula 2 - Importação


```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Acao_pura.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/Acao_pura.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Acao_pura.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/Acao_pura.xlsx", sheet_name="Chat GPT")
df_chatgpt
```

```python
df_principal = df_principal[["Ativo", "Data", "Último (R$)", "Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)":"valor_final", "Var. Dia (%)":"var_dias_pct"}).copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"vat_dias_pct":"var_dias_pct"}).copy()
df_principal
```

```python
df_principal["Var_pct"] = df_principal['var_dias_pct'] / 100
df_principal["Valor_inicial"] = df_principal['valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on="Ativo", right_on="Código", how="left")
df_principal
```

```python
df_principal = df_principal.drop(columns=["Código"])
df_principal
```

```python
df_principal["Variacao_rs"] = (df_principal['valor_final'] - df_principal['Valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format="{:.2f}".format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={"Qtde. Teórica":"qtd_teorica"}).copy()
df_principal
```

```python
df_principal["Resultado"] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on="Ativo", right_on="Ticker", how="left")
df_principal = df_principal.drop(columns=["Ticker_y", "Ticker_x", "Nome_x"])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on="Nome_y", right_on="Nome da empresa", how="left")
df_principal = df_principal.drop(columns=["Nome da empresa"])
df_principal
```

```python
df_principal = df_principal.rename(columns={"Nome_y":"Nome", "Idade (em anos)":"Idade"}).copy()
df_principal
```

```python
df_principal["Idade"] = df_principal["Idade"].apply(lambda x: int(x) if str(x).isnumeric() else None)
df_principal["cat_idade"] = df_principal["Idade"].apply(lambda x: "mais de 100 anos" if x > 100 else ("menos de 50 anos" if x < 50 else "entre 50 e 100 anos"))
df_principal
```

```python
# Calcule os valores solicitados
maior = df_principal['Variacao_rs'].max()
menor = df_principal['Variacao_rs'].min()
media = df_principal['Variacao_rs'].mean()
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Exiba os resultados
print("Maior:", "R$", "{:,.2f}".format(maior))
print("Menor:", "R$", "{:,.2f}".format(menor))
print("Média:", "R$", "{:,.2f}".format(media))
print("Média de quem subiu:", "R$", "{:,.2f}".format(media_subiu))
print("Média de quem desceu:", "R$", "{:,.2f}".format(media_desceu))
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x="Resultado", y="Variacao_rs", text="Variacao_rs", title="Variação Reais por resultado")
fig.show()
```

```python
fig = px.pie(df_analise_segmento, values="Variacao_rs", names="Segmento", title="Variação Reais por resultado")
fig.show()
```
